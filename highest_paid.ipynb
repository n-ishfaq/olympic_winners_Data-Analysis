```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
# plt.style.use("seaborn-pastel")

df = pd.read_excel(r"G:\harvard project\New Microsoft Excel Worksheet.xlsx")
df.head()


```

```python
gb=df.groupby("Athlete")["Gold"].sum()
print(gb)
```

```python
gb.plot(kind="pie",
                 autopct="%.0f%%",  # Percentage format
                 figsize=(10, 10),  # Size of the figure
                 wedgeprops=dict(width=0.4),  # Thickness of the pie
                 pctdistance=0.85)  # Position of percentage labels

# Formatting the plot
plt.ylabel('')  # Removing y-axis label
plt.title("Gold Medal Distribution Among Athletes", fontweight="bold")  # Title

# Display the plot
plt.show()
```
